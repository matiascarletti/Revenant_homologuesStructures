```python
import sys
import os

# If you're using nt data in your working dir

def runBlastdbcmd (blastDatabase, dbIdCodeList, outMfasta):    
    """[create a Multi-Fasta file from id database code list]

    Args:
        blastDatabase ([str]): [pathFile to target blastDatabase]
        idDbCodeList ([str]): [pathFile to a list of idDbCodes \
                                (in commaSeparated fields), \
                                to mapping in the target database]
        outMfasta ([str]): [pathFile to mfasta outFile]
    """
    # Setting command line for run blastdbcmd
    blastdbcmdComandLine = "blastdbcmd -db %s " % blastDatabase + \
                            "-entry $(cat %s) " %  dbIdCodeList + \
                            " > %s" % outMfasta
    print(blastdbcmdComandLine)
    # Runing comand line for run blastdbcmd
    os.system(blastdbcmdComandLine)
    return

rvIdsWithPdbId = "RV4"
# Run function
## Setting paths
pdbDatabase     = "/home/matias/blastDbs/pdbaa/pdbaa"
pdbIdBlaspPdbList = "/home/matias/Projects/2020/Revenant_homologuesStructures/" \
                    "processed_data/2020-06-22/lists/blastpPdb/rvIdsWithPdbId/" \
                    "%s.list" % rvIdsWithPdbId
outPathFolder   = "/home/matias/Projects/2020/Revenant_homologuesStructures/"\
                    "processed_data/2020-06-22/mfastas/" \
                    "homologues/"
pdbMfastaFile       =  "%s_blastpPdbHoms.fasta" % rvIdsWithPdbId
## Creating out path folder
try:
    os.makedirs(outPathFolder)
except FileExistsError:
    print("folder exist")
## Calling funtion
runBlastdbcmd(blastDatabase=pdbDatabase, 
                dbIdCodeList=pdbIdBlaspPdbList, 
                outMfasta=outPathFolder + pdbMfastaFile)
```

```python
#!/usr/bin/env python3
import re
import pandas as pd


def getListOfListFrom(mammothOutFilePath=str):
    """leo el output de mammoth y lo guardo en una lista (cada elemento es una línea del cluster)
    para cada linea uso split y genero listas por cada linea. Esto hace que me quede una lista de listas (cada elemento
    de la lista mammoth ahora es una lista, cuyos elementos son las palabras separadas por coma)

    :param mammothOutFilePath: str
    :return:
    """
    mammothOutFileList = open(mammothOutFilePath).readlines()
    for i in range(len(mammothOutFileList)):
        mammothOutFileList[i] = mammothOutFileList[i].replace(" =", "=")  # .replace("= ", "=") este replace hace que me junte los parámetros con los numeros por el =
        mammothOutFileList[i] = mammothOutFileList[i].split()
    return mammothOutFileList
    # creo el diccionario donde se va a guardar todo


def getMammothRmsd1dAnd3dDictFrom(mammothOutFilePath=str, getPsi1dRmsd=True, getPsi3dRmsd=True):
    """

    :param mammothOutFilePath: the mammoth out file path
    :return:
    """
    mammothOutFileList = getListOfListFrom(mammothOutFilePath)

    mammothRmsd1dAnd3dDict = dict()
    # creo listas vacias para ir guardando las claves y los valores
    claves = []
    valores = []

    for line in mammothOutFileList:
        if getPsi1dRmsd:
            if "PSI(1D)=" in line:
                claves.append("rmsdP" + line[0][1:].lower().replace("=", ""))
                valores.append(line[7])
        if getPsi3dRmsd:
            if "PSI(3D)=" in line:
                claves.append("rmsdP" + line[0][1:].lower().replace("=", ""))
                valores.append(line[7])
    # Por ultimo le voy agregando al diccionario cada clave(cluster) con su respectivo valor(la lista de cadenas pdb)
    for i in range(len(claves)):
        mammothRmsd1dAnd3dDict[claves[i]] = [valores[i]]
    return mammothRmsd1dAnd3dDict


def getMammothRmsd1dAnd3dDfFrom(mammothRmsd1dAnd3dDict=dict):
    mammothRmsd1dAnd3dDf = pd.DataFrame.from_dict(mammothRmsd1dAnd3dDict)
    return mammothRmsd1dAnd3dDf
def getRmsdDataFrameFromRmsdTable(rmsdTablePath):
    rmsdDataframe = pd.read_csv(rmsdTablePath,
                                sep="\t", header=0,
                                index_col=0)
    return rmsdDataframe


def getDataFrameListFormRmsdDataFrame(rmsdDataFrame):
    rmsdDfList = list()
    rmsdDfList.append(rmsdDataFrame)
    return rmsdDfList


def getRmsdDataFramesConcats(rmsdDataFrameList):
    rmsdDfConcats = pd.concat(rmsdDataFrameList)
    return rmsdDfConcats


import os
# for get RvId list with PdbIds
from src.packages.getFasta.getFastaIteratorObjectFromMultifasta import \
    getFastaIteratorObjectFromMultifasta
from src.packages.getFastaHeader.getFastaHeaderFromFasta import \
    getFastaHeaderFromFasta
from src.packages.getFastaHeaderInfo.getFastaHeaderInfoFromFastaHeader import \
    getFastaHeaderInfoFromFastaHeader
from src.packages.getFastaHeaderInfoFilters.getIdEntryFastaWithPdbId import \
    fastaHeaderHavePdbIdEntry
# Para descargar pdb
import urllib
# Para correr mammoth
from src.packages.runMammoth.runMammoth import runMammothForPdbIdEntryRefAndAlt
import os
#
from src.packages.getIoFilePathList.getFilePathList import \
    getFilePathListFromDirectoryPath
from src.packages.getIoFilePathList.getFilePathList import \
    getIdDbListPathFromDirectoryPath

rvIdWithPdbId = "RV4"
# Get path of cluster pdb Id list for each rv id
rvPath = "/home/matias/Projects/2020/Revenant_homologuesStructures" \
            "/outputs/2020-06-22" \
            "/mammoth/nmrModelsWithChainsAndXrayChainsFiles/clustersCodnas" \
            "/%s" % rvIdWithPdbId

rvClusterNameList = os.listdir(rvPath)

for rvClusterName in rvClusterNameList:
    # path infile
    rvClusterPath = "/home/matias/Projects/2020/Revenant_homologuesStructures" \
                    "/outputs/2020-06-22" \
                    "/mammoth/nmrModelsWithChainsAndXrayChainsFiles/clustersCodnas"\
                    "/%s" % rvIdWithPdbId + \
                    "/%s" % rvClusterName

    outMammothFileNameList = os.listdir(rvClusterPath)

    # path outfile
    outPath = "/home/matias/Projects/2020/Revenant_homologuesStructures" \
                "/processed_data/2020-06-22" \
                "/tables" \
                "/mammoth" \
                "/nmrModelsWithChainsAndXrayChainsFiles/clustersCodnas" \
                "/%s" % rvIdWithPdbId + \
                "/%s" % rvClusterName  # outFolderName
    # make outdir
    try:
        os.makedirs(outPath)
    except FileExistsError:
        print("file exist")
    # Cambio a directorio donde estan los outputs para correr el parser
    os.chdir(
        rvClusterPath)  # corro esto o la linea 37? tengo que correr por que me da que no encuentra los mammoth outputs
    for outMammothFileName in outMammothFileNameList:
        outMammothFilePath = os.path.join(rvClusterPath, outMammothFileName)

        mammothRmsd1dAnd3dDict = getMammothRmsd1dAnd3dDictFrom(mammothOutFilePath=outMammothFilePath, getPsi1dRmsd=True, getPsi3dRmsd=True)
        #print(mammothRmsd1dAnd3dDict)
        mammothRmsd1dAnd3dDf = getMammothRmsd1dAnd3dDfFrom(mammothRmsd1dAnd3dDict)
        mammothRmsd1dAnd3dDf["revenantId"] = rvIdWithPdbId
        mammothRmsd1dAnd3dDf["clusterName"] = rvClusterName.replace(".list", "")
        mammothRmsd1dAnd3dDf["pdbIdChainComparison"] = outMammothFileName.replace(".pdb_", "-")\
                                                                            .replace(".pdb.out", "")
        mammothRmsd1dAnd3dDf.to_csv(outPath + "/" +  # path out folder
                    "rmsds_%s" % (outMammothFileName.replace(".pdb_", "-").replace(".pdb.out", "")) +  # path out file name
                        ".tab", sep="\t")
        print(mammothRmsd1dAnd3dDf)

    os.chdir(outPath)
    rmsdTablePathList = os.listdir(outPath)
    rmsdDataFrameList = list()
    for rmsdTablePath in rmsdTablePathList:
        #try:
        rmsdDf = getRmsdDataFrameFromRmsdTable(rmsdTablePath=rmsdTablePath)
        #rmsdDataFrameList = getDataFrameListFormRmsdDataFrame(rmsdDataFrame=rmsdDf)
        rmsdDataFrameList.append(rmsdDf)


    rmsdDfConcats = getRmsdDataFramesConcats(rmsdDataFrameList=rmsdDataFrameList)

    print(rmsdDfConcats.head())
    rmsdDfConcats.to_csv(outPath + "/" +  # path out folder
                            "rmsdConcats" +  # path out file name
                            ".tab", sep="\t")
    try:
        rmsdPsi1dMax = rmsdDfConcats["rmsdPsi(1d)"].max()
        with open(outPath + "/" + "rmsd1dMax.txt", "w") as f:
            f.write(rvClusterName + "\t" + str(rmsdPsi1dMax) + "\t" + outMammothFileName.replace(".pdb_", "-") \
                                                                            .replace(".pdb.out", ""))
    except KeyError:
        print("key error")
        #except FileNotFoundError:
        #    print(rmsdTablePath, " does not exist")
```
